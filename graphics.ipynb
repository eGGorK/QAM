```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('build/QSPK.txt', skiprows=1)
sigma_sq = data[:, 0]
ber = data[:, 1]

plt.figure()
plt.semilogy(sigma_sq, ber, 'o-')
plt.xlabel('SNR(dB)')
plt.ylabel('BER')
plt.grid(True)
plt.title('BER/SNR for QSPK')
plt.show()

data = np.loadtxt('build/QAM16.txt', skiprows=1)
sigma_sq = data[:, 0]
ber = data[:, 1]

plt.figure()
plt.semilogy(sigma_sq, ber, 'o-')
plt.xlabel('SNR(dB)')
plt.ylabel('BER')
plt.grid(True)
plt.title('BER/SNR for QAM16')
plt.show()

data = np.loadtxt('build/QAM64.txt', skiprows=1)
sigma_sq = data[:, 0]
ber = data[:, 1]

plt.figure()
plt.semilogy(sigma_sq, ber, 'o-')
plt.xlabel('SNR(dB)')
plt.ylabel('BER')
plt.grid(True)
plt.title('BER/SNR for QAM64')
plt.show()
```
